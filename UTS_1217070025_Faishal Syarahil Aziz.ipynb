10. Setelah melakukan konversi BGR ke HSV kemudian tampilkan histogram untuk Band Merah/Red Channel, Hijau/Green Channel dan Biru/Blue channel (boleh menggunakan Matplotlib atau cv2.imshow) (Poin: 10)




```python
#Tuliskan kode program disini

# 1217070025
# Faishal Syarahil Aziz

import cv2
import matplotlib.pyplot as plt

img = cv2.imread('images.jpg')

b, g, r = cv2.split(img)

histogram_b = cv2.calcHist([b], [0], None, [256], [0, 256])
histogram_g = cv2.calcHist([g], [0], None, [256], [0, 256])
histogram_r = cv2.calcHist([r], [0], None, [256], [0, 256])

plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
plt.plot(histogram_r, color='red')
plt.title('Histogram Red Channel')

plt.subplot(1, 3, 2)
plt.plot(histogram_g, color='green')
plt.title('Histogram Green Channel')

plt.subplot(1, 3, 3)
plt.plot(histogram_b, color='blue')
plt.title('Histogram Blue Channel')

plt.show()
```

12. Menaikan brighness dengan menambahkan nilai konstanta = 100. Tampilkan gambar sebelum brightness dan sesudah (poin: 10)

```python
#Tuliskan kode program disini

# 1217070025
# Faishal Syarahil Aziz

import cv2
import matplotlib.pyplot as plt

img = cv2.imread('images.jpg')

hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

h, s, v = cv2.split(hsv_img)

brightness_increase = 100
imgbrigh = cv2.add(v, brightness_increase)

imgbrigh = cv2.min(imgbrigh, 255)

hsv_brigh = cv2.merge([h, s, imgbrigh])

rgb_brigh = cv2.cvtColor(hsv_brigh, cv2.COLOR_HSV2BGR)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Gambar Asli')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(rgb_brigh, cv2.COLOR_BGR2RGB))
plt.title(f'Gambar setelah Brightness')
plt.axis('off')

plt.show()
```

14. Normalisasikan gambar dan tampilkan historgramnya (poin 5)

```python
#Tuliskan kode program disini

# 1217070025
# Faishal Syarahil Aziz

import cv2
import matplotlib.pyplot as plt

img = cv2.imread('images.jpg', cv2.IMREAD_GRAYSCALE)

normalized_image = cv2.normalize(img, None, 0, 255, cv2.NORM_MINMAX)

hist = cv2.calcHist([normalized_image], [0], None, [256], [0, 256])

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(normalized_image, cmap='gray')
plt.title('Gambar Normalisasi')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.plot(hist, color='black')
plt.title('Histogram Gambar Normalisasi')

plt.show()
```
